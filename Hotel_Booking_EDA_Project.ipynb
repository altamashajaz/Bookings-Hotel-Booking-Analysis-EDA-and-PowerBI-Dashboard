## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
# Let's Start with data cleaning.
# Remove duplicate rows from the hotel_booking DataFrame, modifying it in place
hotel_booking.drop_duplicates(inplace=True)
```

```python
# Dealing with missing values.
# Dropping Company column because it contains too many null value.
hotel_booking = hotel_booking.drop('company', axis = 1)
```

```python
# Fill missing values in the 'agent' column with 'N/A' since agent column holds Agent ID it is best replacing null with "N/A"
hotel_booking['agent'] = hotel_booking['agent'].fillna('N/A')
```

```python
# Fill missing values in the 'country' column with 'Not Specified' cause "country" is a categorical data and for null using "Not Specified make most sense."
hotel_booking['country'] = hotel_booking['country'].fillna('Not Specified')
```

```python
# Drop rows with missing values in the 'children' column since it is very few it make sense.
hotel_booking.dropna(subset=['children'], inplace=True)
```

```python
# Calculate total stay by summing up the number of weekend nights and week nights
hotel_booking['Total_stay'] = hotel_booking['stays_in_weekend_nights'] + hotel_booking['stays_in_week_nights']
```

```python
# Calculate total number of kids by summing up the number of babies and children
hotel_booking['Total_kids'] = hotel_booking['babies'] + hotel_booking['children']
```

```python
# Calculate total cost by multiplying the average daily rate (adr) by the total stay
hotel_booking['total_cost'] = hotel_booking['adr']*hotel_booking['Total_stay']
```

```python
# Create a new date cloumn.
hotel_booking['arrival_date'] = pd.to_datetime(hotel_booking['arrival_date_year'].astype(str) + '-' + hotel_booking['arrival_date_month'] + '-' + hotel_booking['arrival_date_day_of_month'].astype(str))
```

```python
# Get a list of column names in the hotel_booking DataFrame
hotel_booking.columns.to_list()
```

#### **Cancellation Rate by Deposit Types**

```python
# Count the total bookings for each deposit type
total_booking_dep = hotel_booking['deposit_type'].value_counts().reset_index(name='total_booking').set_index('index')
# Group by deposit type and calculate the total number of cancellations
deposit_type_on_canceled = hotel_booking.groupby('deposit_type')['is_canceled'].sum().reset_index().set_index('deposit_type')
# Join the total bookings and cancellations by deposit type
combined = deposit_type_on_canceled.join(total_booking_dep, how='inner').reset_index()
# Calculate cancellation rate for each deposit type
combined['cancel_rate %'] = round((combined['is_canceled'] / combined['total_booking']) * 100, 2)
# Display the combined DataFrame
print(combined)
```

```python
# Chart - 16 visualization code
plt.figure(figsize=(8, 6))
sns.barplot(data=combined, x='deposit_type',  y='cancel_rate %', color='skyblue')
plt.title('Cancellation Rate by Deposit Type')
plt.xlabel('Deposit Type')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.ylabel('Cancellation Rate (%)')
plt.ylim(0, 100)  # Set y-axis limits to show percentage
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
